```python
import pandas as pd
df = pd.read_csv(file_path)
df.columns = df.columns.str.lower().str.strip()

useful_col = ["title", "genre", "overview", "release_date", "id"]
df = df[[col for col in useful_col if col in df.columns]]

df.dropna(subset="title", inplace=True)
df.drop_duplicates(subset="title", inplace=True)
df.reset_index(drop=True, inplace=True)
df.isna().sum()
```

```python
import time
import requests
from tqdm import tqdm
api_key = "b7e4526c"
limit = 300
titles = df["title"].head(limit)
movie_data = []

for title in tqdm(titles):
    try:
        url = f"http://www.omdbapi.com/?t={title}&apikey={api_key}"
        response = requests.get(url)
        data = response.json()
        if data.get('Response') == 'True':
            movie_data.append({
                'title': data.get('Title'),
                'imdb_rating': data.get('imdbRating'),
                'runtime': data.get('Runtime'),
                'director': data.get('Director'),
                'actors': data.get('Actors'),
                'plot_omdb': data.get('Plot')
            })

        time.sleep(0.3)
    except Exception as e:
      print(f"Error fetching {title}; {e}")
```

```python
import requests
from bs4 import BeautifulSoup

def check_watch_provider(imdb_id):
    try:
        url = f"https://www.imdb.com/title/{imdb_id}/"
        headers = {"User-Agent": "Mozilla/5.0"}
        reponse = requests.get(url, headers=headers, timeout=10)

        if reponse.status_code != 200:
            print(f"IMDB page not found")
            return None
        
        soup = BeautifulSoup(reponse.text, "html.parser")
        text = soup.get_text(" ").lower()

        ott_keywords = {
            "Netflix": ["netflix"],
            'Amazon Prime Video': ['primevideo','amazon'],
            'Disney+ hotstar': ['disney', 'hotstar'],
            'Zee5': ['zee5'],
            'Youtube': ['youtube', 'googleplay'],
        }

        found = [name for name, keys in ott_keywords.items() if any(k in text for k in keys)]
        if found:
            print(f"found providers for {imdb_id}: {','.join(found)}")

            return found
        else:
            print(f"No providers found")
            return None
    except Exception as e:
        print(f"error fetching imdb page: {e}")
        return None
    


test_ids = ["tt1375666",
            "tt11866324",
            "tt1187043"]

for mid in test_ids:
    check_watch_provider(mid)
```

```python
import re
def get_imdb_rating(imdb_id):
    if imdb_id is None:
        return None
    
    url = f"https://www.imdb.com/title/{imdb_id}/"

    headers = {"User-Agent": "Mozilla/5.0"}

    try:
        r = requests.get(url, headers=headers,timeout=10)

        if r.status_code != 200:
            print("Rating error:", r.status_code)
            return None
        
    except:
        return None
    
    soup = BeautifulSoup(r.text, "html.parser")

    text = soup.get_text(" ", strip=True)

    match = re.search(r"(\d\.\d)\s*/\s*10", text)
    if match:
        return match.group(1)
    
    return None

test_ids = ["tt1375666",
            "tt11866324",
            "tt4154796"]

for mid in test_ids:
    print(get_imdb_rating(mid))
```

```python
import requests


def get_watch_provider(title, year=None):
    url = f"https://apis.justwatch.com/content//titles/en_IN/popular"

    payload = {
        "page_size": 5,
        "page": 1,
        "query": title
    }

    headers = {
        "User-Agent": "Mozilla/5.0",
        "Content-Type": "applicaation/json"
    }

    response = requests.post(url, json=payload,headers=headers)
    print("Status:", response.status_code)


    if response.status_code != 200:
        print("Error:", response.status_code)
        return None
    try:
        data = response.json()
    except:
        print("JSON ERROR")
        return None
    


    if "items" not in data or len(data["items"]) ==0:
        return None
    
    item = data["items"][0]
    offers = item.get("offers", [])
    if not offers:
        return []
    


    providers = set()

    for off in offers:
        provider = off.get("package_short_name")
        if provider:
            providers.add(provider)


    return list(providers)


    

print(get_watch_provider("Inception"))

    



```
