```python
import os
path = r"./datasets/archive (2)"
file_path = os.path.join(path,"movies_metadata.csv")
print("file_path",file_path)
print("Exists",os.path.exists(file_path))
```

```python
import pandas as pd
df = pd.read_csv(file_path)
df.columns = df.columns.str.lower().str.strip()

useful_col = ["title", "genre", "overview", "release_date", "id"]
df = df[[col for col in useful_col if col in df.columns]]

df.dropna(subset="title", inplace=True)
df.drop_duplicates(subset="title", inplace=True)
df.reset_index(drop=True, inplace=True)
df.isna().sum()
```

```python
import time
import requests
from tqdm import tqdm
api_key = "b7e4526c"
limit = 300
titles = df["title"].head(limit)
movie_data = []

for title in tqdm(titles):
    try:
        url = f"http://www.omdbapi.com/?t={title}&apikey={api_key}"
        response = requests.get(url)
        data = response.json()
        if data.get('Response') == 'True':
            movie_data.append({
                'title': data.get('Title'),
                'imdb_rating': data.get('imdbRating'),
                'runtime': data.get('Runtime'),
                'director': data.get('Director'),
                'actors': data.get('Actors'),
                'plot_omdb': data.get('Plot')
            })

        time.sleep(0.3)
    except Exception as e:
      print(f"Error fetching {title}; {e}")
```
